# Processing OSISAF data

### Set up
#### Packages

```python
import numpy as np
import xarray as xr
import pandas as pd
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
import datetime as dt
from datetime import timedelta
xr.set_options(display_expand_data=False);
```

#### Filepaths & name variables

```python
## File name
filename1 = 'osisaf_nh_sie_monthly.nc'

## Filepaths
path_to_arch = "/glade/work/glydia/processed_OSISAF_data/"
comp = 'ice'
var_ind = 0

# Variables
var_list = {'ice': [['sie','aice']]}
var = var_list[comp][var_ind][0]
var2 = var_list[comp][var_ind][1]

# Extensions
h_ext = {'ice': ['.h.']}

path_to_outdata = '/glade/work/glydia/processed_OSISAF_data/'
```

### Load & modify data
#### Control data

```python
%%time

## Load data
# Open csvs
ds = xr.open_dataset(path_to_arch+filename1)
dsv = ds[var]
dsv = dsv.rename(var2)

dsv.to_netcdf(path_to_outdata+'OSISAF'+h_ext[comp][0]+var2+'.195001-202412.'+'nc', 
                            format='NETCDF4',encoding={var2: {"zlib": True, "complevel": 1}})
print('wrote data to disk')
```
